```python
df = pd.read_csv('Data/comments.csv')
df.head()
```

```python
# Set 'Category' as index
df.set_index('Category', inplace=True)


# Use the 'inferno' colormap and normalise the color values
colors = plt.cm.Pastel1(np.linspace(0, 1, len(df)))  # Normalise color values between 0 and 1

# Plot pie chart
df['NumberOfComments'].plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6),
                            textprops={'fontsize': 14},  
                            colors=colors)

# Set custom y-label
plt.ylabel('')


# Optional: Set plot title
plt.title('')

plt.savefig('Figure/commentsCa.png', bbox_inches='tight')
```
