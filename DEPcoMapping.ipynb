```python
df = pd.read_csv('Data/site.csv') 
df.head()
```

```python
plt.figure(figsize=(16, 9))

# Plot the bar chart
plt.bar(df['name'], df['min/hec'], color = 'steelblue', edgecolor='black')

# Add labels and title
plt.xlabel('',fontsize=15)
plt.ylabel('Minutes/Person/Hectare',fontsize=15)
plt.title('')
plt.xticks(fontsize=15, rotation=45)
plt.yticks(fontsize=12)
plt.savefig('Figure/timeNormalised.png', bbox_inches='tight')
```

```python
df = pd.read_csv('Data/comments.csv')
df.head()
```

```python
# Set 'Category' as index
df.set_index('Category', inplace=True)


# Use the 'inferno' colormap and normalise the color values
colors = plt.cm.Pastel1(np.linspace(0, 1, len(df)))  # Normalise color values between 0 and 1

# Plot pie chart
df['NumberOfComments'].plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6),
                            textprops={'fontsize': 14},  
                            colors=colors)

# Set custom y-label
plt.ylabel('')


# Optional: Set plot title
plt.title('')

plt.savefig('Figure/commentsCa.png', bbox_inches='tight')
```
